```python
# Import pandas library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Read the CSV file
df = pd.read_csv("Weather.csv")

# Display the first 5 rows
print("First 5 Rows")
print(df.head())

# Show column names
print("\nColumn Names")
print(df.columns)

```

```python
# 1. Find unique wind speeds
print("\n1. Unique Wind Speeds")
print(df["Wind Speed_km/h"].unique())
```

```python
# 2. Find the number of times when the weather is exactly Clear
print("\n2. Weather is Clear")
print(df[df["Weather"] == "Clear"].count())
```

```python
# 3. Find all records where wind speed is exactly 4
print("\n3. Wind Speed = 4")
print(df[df["Wind Speed_km/h"] == 4])
```

```python
# 4. Find the mean visibility
print("\n4. Mean Visibility")
print(df["Visibility_km"].mean())
```

```python
# 5. Find the standard deviation of Pressure
print("\n5. Standard Deviation of Pressure")
print(df["Press_kPa"].std())
```

```python
# 6. Find all instances when Snow was recorded
print("\n6. Snow Records")
print(df[df["Weather"].str.contains("Snow")])
```

```python
# 7. Show all records when the weather is Fog
print("\n7. Fog Records")
print(df[df["Weather"] == "Fog"])
```

```python
# 8. Find the mean value of each column against each weather condition
print("\n8. Mean Value of Each Column by Weather")
print(df.groupby("Weather").mean(numeric_only=True))
```

```python
# 9. Find all instances where Weather is Clear OR Visibility is above 40
print("\n9. Weather is Clear OR Visibility > 40")
print(df[(df["Weather"] == "Clear") | (df["Visibility_km"] > 40)])
```

```python
# 10. Find all instances where Weather is Clear AND Relative Humidity is above 50
print("\n10. Weather is Clear AND Relative Humidity > 50")
print(df[(df["Weather"] == "Clear") & (df["Rel Hum_%"] > 50)])
```

```python
# Read the CSV file
df = pd.read_csv("weather.csv")

# Rename the column
df.rename(columns={"Weather": "Weather Condition"}, inplace=True)

# Check the new column names
print(df.columns)
```

```python
weather_count = df["Weather Condition"].value_counts()

weather_count.plot(kind="bar")

plt.title("Weather Condition Count")
plt.xlabel("Weather")
plt.ylabel("Count")
plt.show()
```

```python
weather_count = df["Weather Condition"].value_counts()

weather_count.plot(kind="pie", autopct="%1.1f%%")

plt.title("Weather Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.plot(df["Temp_C"])

plt.title("Temperature")
plt.xlabel("Record Number")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.hist(df["Wind Speed_km/h"], bins=10)

plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Frequency")
plt.show()
```

```python
group=df.groupby("Weather Condition").mean(numeric_only=True)

group.plot(kind="bar",
           figsize=(16,8),
           width=0.8)

plt.title("Average Weather Parameters by Weather Condition")
plt.xlabel("Weather")
plt.ylabel("Mean Value")
plt.xticks(rotation=90)

plt.grid(axis='y',alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.boxplot(df["Press_kPa"])

plt.title("Pressure Box Plot")
plt.ylabel("Pressure (kPa)")
plt.show()
```

```python
plt.plot(df["Rel Hum_%"])

plt.title("Relative Humidity")
plt.xlabel("Record Number")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
plt.hist(df["Visibility_km"], bins=10)

plt.title("Visibility Distribution")
plt.xlabel("Visibility (km)")
plt.ylabel("Frequency")
plt.show()
```
